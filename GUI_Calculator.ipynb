```python
from tkinter import *
```

```python
root = Tk()
root.configure(bg="#55342b")
root.title("Handsome Calculator")

#Entry
e = Entry(root)

#Functions
def butt_pressed_0():
    return e.insert("end", "0")

def butt_pressed_1():
    return e.insert("end", "1")

def butt_pressed_2():
    return e.insert("end", "2")

def butt_pressed_3():
    return e.insert("end", "3")

def butt_pressed_4():
    return e.insert("end", "4")

def butt_pressed_5():
    return e.insert("end", "5")

def butt_pressed_6():
    return e.insert("end", "6")

def butt_pressed_7():
    return e.insert("end", "7")

def butt_pressed_8():
    return e.insert("end", "8")

def butt_pressed_9():
    return e.insert("end", "9")

def butt_pressed_sum():
    global n1
    n1 = e.get()
    global k
    k = "sum"
    return e.delete(0, "end")

def butt_pressed_sub():
    global n1
    n1 = e.get()
    global k
    k = "sub"
    return e.delete(0, "end")

def butt_pressed_mult():
    global n1
    n1 = e.get()
    global k
    k = "mult"
    return e.delete(0, "end")

def butt_pressed_div():
    global n1
    n1 = e.get()
    global k
    k = "div"
    return e.delete(0, "end")

def butt_pressed_equal():
    n2 = e.get()
    e.delete(0, "end")
    if k == "sum":
        return e.insert(0, int(n1)+int(n2))
    if k == "sub":
        return e.insert(0, int(n1)-int(n2))
    if k == "mult":
        return e.insert(0, int(n1)*int(n2))
    if k == "div":
        return e.insert(0, int(n1)/int(n2))

def butt_pressed_clear():
    return e.delete(0, "end")

#Buttons
button0 = Button(root, text="0", padx=35, pady=35, command=butt_pressed_0)
button1 = Button(root, text="1", padx=35, pady=35, command=butt_pressed_1)
button2 = Button(root, text="2", padx=35, pady=35, command=butt_pressed_2)
button3 = Button(root, text="3", padx=35, pady=35, command=butt_pressed_3)
button4 = Button(root, text="4", padx=35, pady=35, command=butt_pressed_4)
button5 = Button(root, text="5", padx=35, pady=35, command=butt_pressed_5)
button6 = Button(root, text="6", padx=35, pady=35, command=butt_pressed_6)
button7 = Button(root, text="7", padx=35, pady=35, command=butt_pressed_7)
button8 = Button(root, text="8", padx=35, pady=35, command=butt_pressed_8)
button9 = Button(root, text="9", padx=35, pady=35, command=butt_pressed_9)
butt_sum = Button(root, text="+", padx=35, pady=35, command=butt_pressed_sum)
butt_sub= Button(root, text="-", padx=35, pady=35, command=butt_pressed_sub)
butt_div = Button(root, text="/", padx=35, pady=35, command=butt_pressed_div)
butt_mult= Button(root, text="x", padx=35, pady=35, command=butt_pressed_mult)
butt_equal = Button(root, text="=", padx=35, pady=35, command=butt_pressed_equal)
butt_clean = Button(root, text="C", padx=35, pady=35, command=butt_pressed_clear)

#Buttons grid
#sticky='nesw' will allow the buttons to expand with their respective row and column sizes.
e.grid(row=0, column=0, columnspan=3, sticky='nesw', padx=25,pady=25,ipady=5)
button0.grid(row=4, column=0)
button1.grid(row=1, column=0, sticky='nesw')
button2.grid(row=1, column=1, sticky='nesw')
button3.grid(row=1, column=2, sticky='nesw')
button4.grid(row=2, column=0, sticky='nesw')
button5.grid(row=2, column=1, sticky='nesw')
button6.grid(row=2, column=2, sticky='nesw')
button7.grid(row=3, column=0, sticky='nesw')
button8.grid(row=3, column=1, sticky='nesw')
button9.grid(row=3, column=2, sticky='nesw')
butt_sum.grid(row=5, column=0, sticky='nesw')
butt_sub.grid(row=5, column=1, sticky='nesw')
butt_div.grid(row=6, column=1, sticky='nesw')
butt_mult.grid(row=6, column=0, sticky='nesw')
butt_equal.grid(row=4, column=1, sticky='nesw', columnspan=2)
butt_clean.grid(row=5, column=2, sticky='nesw', rowspan=2)

root.mainloop()
```
